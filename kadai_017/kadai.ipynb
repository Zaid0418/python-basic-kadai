```python
# ------------------------------
# Humanクラスの定義
# ------------------------------
class Human:
    def __init__(self, name, age):
        # インスタンス属性（スコープ：インスタンス内）
        self.name = name
        self.age = age

    def check_adult(self):
        # 条件分岐を使用
        if self.age >= 20:
            print(f"{self.name} さん（{self.age}歳）は大人です。")
        else:
            print(f"{self.name} さん（{self.age}歳）は大人ではありません。")


# ------------------------------
# Humanインスタンスを作成する関数
# ------------------------------
def create_humans():
    # ディクショナリを使用してデータ管理
    data_list = [
        {"name": "太郎", "age": 25},
        {"name": "花子", "age": 18},
        {"name": "次郎", "age": 30},
    ]

    humans = []  # リスト（スコープ：関数内）

    # 繰り返し処理でHumanインスタンスを生成
    for data in data_list:
        human = Human(data["name"], data["age"])
        humans.append(human)

    return humans


# ------------------------------
# メイン処理（スコープ：グローバル）
# ------------------------------
def main():
    human_list = create_humans()

    # リストの要素数だけ check_adult を実行
    for human in human_list:
        human.check_adult()


# 実行
main()

```
