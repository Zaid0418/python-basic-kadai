```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(self.name + "is_adult.")
        else:
            print(self.name + "is not adult.")

person = Human("太郎", 35)
person.check_adult()
```
