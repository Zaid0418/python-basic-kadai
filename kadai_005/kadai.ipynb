```python
# 台形の寸法（cm）

upper = 10  # 上辺
lower = 10  # 下辺
height = 5  # 高さ

# 台形の面積を計算（㎠）

area_cm2 = (upper + lower) * height / 2

print(f"台形の面積は{area_cm2}㎠です。")
```
