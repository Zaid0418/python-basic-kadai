```python
import pandas as pd

df_2022 = pd.read_excel('2022_年間売上表.xlsx')
df_2023 = pd.read_excel('2023_年間売上表.xlsx')

df_2022 = df_2022[['商品', '金額（千円）']]
df_2023 = df_2023[['商品', '金額（千円）']]

df_2022['売上年'] = 2022
df_2023['売上年'] = 2023

df = pd.concat([df_2022, df_2023], ignore_index=True)

df = df[['商品', '売上年', '金額（千円）']]

df = df.sort_values(by=['商品', '売上年'])

print(df)
```

```python
result = (
    df
    .groupby(['売上年', '商品'], as_index=False)['金額（千円）']
    .sum()
)

print(result)
```

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

df_2022 = pd.read_excel('2022_年間売上表.xlsx')
df_2023 = pd.read_excel('2023_年間売上表.xlsx')

df_2022.columns = ['売上年', '商品', '金額（千円）']
df_2023.columns = ['売上年', '商品', '金額（千円）']

df = pd.concat([df_2022, df_2023], ignore_index=True)

result = (
    df
    .groupby(['売上年', '商品'], as_index=False)['金額（千円）']
    .sum()
    .sort_values(['商品', '売上年'])
    .reset_index(drop=True)
)

result = result[['商品', '売上年', '金額（千円）']]

file_name = '売上集計表.xlsx'
result.to_excel(file_name, index=False)

wb = load_workbook(file_name)
ws = wb.active

header_fill = PatternFill(
    fill_type='solid',
    start_color='F2F2F2',
    end_color='F2F2F2'
)

for cell in ws[1]:
    cell.fill = header_fill

wb.save(file_name)
```
