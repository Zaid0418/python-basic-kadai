```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def printinfo(self):
        print(f"名前：{self.name}, 年齢：{self.age}")

person = Human("name", 10)
person.printinfo()
```

```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(self.name + "is_adult.")
        else:
            print(self.name + "is not adult.")

person = Human("太郎", 35)
person.check_adult()
```

```python
import requests
from bs4 import BeautifulSoup

# 取得したいページURL
url = "https://www.rentalmarket-houjin.jp/"  # ←ここに対象URLを入れる

# Webページを取得
response = requests.get(url)
response.encoding = response.apparent_encoding

# HTMLを解析
soup = BeautifulSoup(response.text, "html.parser")

# 価格らしき部分を抽出（例として class="price" の要素を検索）
prices = soup.find_all(class_="price")

# 結果を表示
for p in prices:
    print("価格:", p.get_text(strip=True))
```
